# 1. Importing the necessary libraries :

```python
import cv2
import pytesseract
```

# 2. Text recognition from an image :

* And now we are going to use Python-tesseract which is an optical character recognition (OCR) tool for python that recognizes the text embedded in images.
* To install it go to this link : https://github.com/UB-Mannheim/tesseract/wiki

```python
pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"
```

* Reading the picture that we are going to work on

```python
img=cv2.imread("C:/Users/khaoula/Desktop/Text detection/text_picture.png")
```

```python
from IPython.display import Image
Image(filename='C:/Users/khaoula/Desktop/Text detection/text_picture.png') 
```

* Pytesseract doesn't accept .png format so we have to convert it first into RGB 

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow('RGB_Image',img)
cv2.waitKey(0)
```

* And now we are going to extract text from the picture using tesseract image_to_string function.

```python
print(pytesseract.image_to_string(img))
```

# 3.  Characters Detection : 

* And now let's see the position of each character : x , y , width and height and draw a bounding box around the character 

```python
hImg,wImg,_ = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    b=b.split(' ')
    x,y,w,h=int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(img,(x,hImg-y),(w,hImg-h),(0,0,255),1)
cv2.imshow('RGB_Image',img)
cv2.waitKey(0)
```

* And this is the result that we get :

```python
Image(filename='C:/Users/khaoula/Desktop/Text detection/result.png') 
```

# 4. Words Detection :

```python
hImg,wImg,_ = img.shape
boxes = pytesseract.image_to_data(img)
for x,b in enumerate(boxes.splitlines()):
    if x!=0:
      b = b.split()
      print(b)
      if len(b)==12:
          x,y,w,h=int(b[6]),int(b[7]),int(b[8]),int(b[9])
          cv2.rectangle(img,(x,y),(w+x,h+y),(0,0,255),2)
cv2.imshow('RGB_Image',img)
cv2.waitKey(0)
```

* And this is the result that we get :

```python
Image(filename='C:/Users/khaoula/Desktop/Text detection/second_result.png') 
```

# 5. Only Digits Detection :

```python
hImg,wImg,_ = img.shape
cong = r'--oem 3 --psm 6 outputbase digits'
boxes = pytesseract.image_to_data(img,config=cong)
for x,b in enumerate(boxes.splitlines()):
    if x!=0:
      b = b.split()
      print(b)
      if len(b)==12:
          x,y,w,h=int(b[6]),int(b[7]),int(b[8]),int(b[9])
          cv2.rectangle(img,(x,y),(w+x,h+y),(255,0,0),2)
cv2.imshow('RGB_Image',img)
cv2.waitKey(0)
```
